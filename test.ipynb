```python
import numpy as np
from sklearn.datasets import make_blobs
import matplotlib.pyplot as plt
from ClusteringClass import Clustering


k = 5
random_choice = 11

# Generate a test set with 3 clusters
np.random.seed(random_choice)
X, y = make_blobs(n_samples=500, centers=k, cluster_std=1.0)

cmap=plt.get_cmap("jet", k)
plt.scatter(X[:, 0], X[:, 1], c=y, cmap=cmap)
plt.title('True clusters')
plt.show()
```

```python
kmeans = Clustering(k=k, algorithm_variant='kmeans')
kmeans.fit(X, random_choice=random_choice)

kmeanspp = Clustering(k=k, algorithm_variant='kmeans++')
kmeanspp.fit(X, random_choice=random_choice)

kmedoids = Clustering(k=k, algorithm_variant='kmedoids')
kmedoids.fit(X, random_choice=random_choice)

kmedoidspp = Clustering(k=k, algorithm_variant='kmedoids++')
kmedoidspp.fit(X, random_choice=random_choice)

cmeans = Clustering(k=k, algorithm_variant='cmeans')
cmeans.fit(X, random_choice=random_choice)
```

```python
import numpy as np
import pandas as pd

# Define the data
algorithm_names = [kmeans.algorithm_variant, kmeanspp.algorithm_variant, kmedoids.algorithm_variant, kmedoidspp.algorithm_variant, cmeans.algorithm_variant]
running_times = [kmeans.TIME, kmeanspp.TIME, kmedoids.TIME, kmedoidspp.TIME, cmeans.TIME]

# Create a pandas DataFrame to hold the data
data = {'Algorithm': algorithm_names, 'Running Time (s)': running_times}
df = pd.DataFrame(data)

# Format the DataFrame for display as a table
df = df.style.set_table_styles(
    [{'selector': 'th', 'props': [('background-color', 'black'),
                                  ('color', 'white'),
                                  ('font-weight', 'bold')]},
     {'selector': 'td', 'props': [('border', '1px solid black'),
                                  ('text-align', 'center'),
                                  ('padding', '5px')]},
     {'selector': 'tr:nth-child(even)', 'props': [('background-color', '#f2f2f2')]},
     {'selector': 'tr:hover', 'props': [('background-color', 'lightgrey')]}
])

# Display the table
display(df)

# Convert the formatted DataFrame to a string
table_string = df.to_string()

# Write the string to a text file
with open('table.txt', 'w') as f:
    f.write(table_string)
```

```python
kmeans.plot(X, cool=True)
kmeanspp.plot(X, cool=True)
kmedoids.plot(X)
kmedoidspp.plot(X)
cmeans.plot(X, cool=True)
```
